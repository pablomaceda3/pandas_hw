```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools

# Slice column of school names from school data, and count 
# Check if matches schools_complete.
num_schools = school_data['school_name'].count()
num_schools
```

```python
# Calculate the total number of students

# Slice a list of students from the master dataframe. Not unique, two students can have
# the same name and be different people.
# Check if matches students_complete. Has to be different from schools_complete check.
num_students = school_data_complete['student_name'].count()
num_students
```

```python
# Calculate the total budget

# This data has to be grabbed from school_complete dataframe. 
# Too many duplicates to parse from master dataframe. 
# Slice the 'budget' column and add all elements in it. 
total_budget = school_data['budget'].sum()
total_budget
```

```python
# Calculate the average math score

# Slice math_score column from master data frame and .mean() it.
# Round for legibility's sake.
avg_math_score = round(school_data_complete['math_score'].mean(), 2)
avg_math_score
```

```python

# Calculate the average reading score

# Do same as above.
avg_reading_score = round(school_data_complete['reading_score'].mean(), 2)
avg_reading_score
```

```python

# Calculate the percentage of students with a passing math score (70 or greater)

# Here's where it gets interesting.
# First, slice out where passing math score is 70 =>. 
# Then, calculate the percentage by dividing by total students.
stu_pass_math = school_data_complete.loc[school_data_complete['math_score'] >= 70, :]
num_stu_pass_math = stu_pass_math['student_name'].count()

pct_stu_pass_math = round(((num_stu_pass_math/num_students) * 100), 2)
pct_stu_pass_math
```

```python

# Calculate the percentage of students with a passing reading score (70 or greater)

# First, slice out where passing reading score is 70 =>. 
# Then, calculate the percentage by dividing by total students.
stu_pass_reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70, :]
num_stu_pass_reading = stu_pass_reading['student_name'].count()

pct_stu_pass_reading = round(((num_stu_pass_reading/num_students) * 100), 2)
pct_stu_pass_reading
```

```python
# Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

# Pretty self-explanatory.
ov_pass_rate = round(((pct_stu_pass_reading + pct_stu_pass_math)/2), 2)
ov_pass_rate
```

```python

# Create a dataframe to hold the above results

district_summary = pd.DataFrame(
    {'Total Schools': [num_schools],
     'Total Students': [num_students],
     'Total Budget': [total_budget],
     'Average Math Score': [avg_math_score],
     'Average Reading Score': [avg_reading_score],
     '% Passing Math': [pct_stu_pass_math],
     '% Passing Reading': [pct_stu_pass_reading],
     '% Overall Pass Rate': [ov_pass_rate]}
)
```

```python

# Optional: give the displayed data cleaner formatting

# Format numbers greater than 1000 by inserting comments. 
# Format budget with dollar sign

district_summary['Total Students'] = district_summary['Total Students'].astype(int).apply(lambda x: "{:,}".format(x))
district_summary['Total Budget'] = district_summary['Total Budget'].astype(int).map("${:,.2f}".format)
district_summary
```

```python
# % Passing Math
stu_pass_math = school_data_complete.loc[school_data_complete['math_score'] >= 70, :]
stu_pass_math_school = stu_pass_math.groupby(['school_name'])
num_stu_pass_math_school = stu_pass_math_school.count()
num_stu_pass_math_school = num_stu_pass_math_school['size']

pct_stu_pass_math_school = round((num_stu_pass_math_school/tot_stu_per_school * 100), 2)
pct_stu_pass_math_school
```

```python
# % Passing Reading
stu_pass_reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70, :]
stu_pass_reading_school = stu_pass_reading.groupby(['school_name'])
num_stu_pass_reading_school = stu_pass_reading_school.count()
num_stu_pass_reading_school = num_stu_pass_reading_school['size']

pct_stu_pass_reading_school = round((num_stu_pass_reading_school/tot_stu_per_school * 100), 2)
pct_stu_pass_reading_school
```

```python
# # Overall Passing Rate (Average of the above two)
ov_pass_rate_per_school = round(((pct_stu_pass_math_school+pct_stu_pass_reading_school)/2), 2)
ov_pass_rate_per_school
```

```python
# Create a dataframe to hold the above results

# for types in school_type:
#     school_type = school_type.astype('str')

school_summary = pd.DataFrame(
    {'School Type': school_type,
     'Total Students': tot_stu_per_school,
     'Total School Budget': tot_budget_per_school,
     'Per Student Budget': per_stu_budget,
     'Average Math Score': avg_math_score_per_school,
     'Average Reading Score': avg_reading_score_per_school,
     '% Passing Math': pct_stu_pass_math_school,
     '% Passing Reading': pct_stu_pass_reading_school,
     '% Overall Pass Rate': ov_pass_rate_per_school}
)

school_summary
```

## Math Scores by Grade

```python
# Could use a function here.
# If revisiting notes/HW for practice, ask 
# TAs how best to make this section iterable
# by using a function within a `for` loop.

ninth_grade = school_data_complete.loc[school_data_complete['grade'] == '9th', :]
tenth_grade = school_data_complete.loc[school_data_complete['grade'] == '10th', :]
eleventh_grade = school_data_complete.loc[school_data_complete['grade'] == '11th', :]
twelfth_grade = school_data_complete.loc[school_data_complete['grade'] == '12th', :]

ninth_grade_avg_math = ninth_grade['math_score'].mean().round(decimals=2)
tenth_grade_avg_math = tenth_grade['math_score'].mean().round(decimals=2)
eleventh_grade_avg_math = eleventh_grade['math_score'].mean().round(decimals=2)
twelfth_grade_avg_math = twelfth_grade['math_score'].mean().round(decimals=2)

grades = ['9th Grade', '10th Grade', '11th Grade', '12th Grade']

math_score_grade = (
    {'Math Score Average': [ninth_grade_avg_math, tenth_grade_avg_math,
                            eleventh_grade_avg_math, twelfth_grade_avg_math],
     'Grade': grades
    }
)                     
                    
math_score_grade = pd.DataFrame(math_score_grade)
math_score_grade.set_index('Grade', inplace=True)
math_score_grade
```
